### **2. What is Bayes’ Theorem? What is posterior and  likelihood?**

**Bayes' Theorem** is a mathematical formula used in statistics and probability theory to calculate the probability of an event occurring, given some prior knowledge or evidence. It was named after the Reverend Thomas Bayes, an 18th-century British statistician.

The theorem can be stated as follows:

P(A|B) = P(B|A) * P(A) / P(B)

where P(A|B) is the probability of event A occurring given that event B has occurred, P(B|A) is the probability of event B occurring given that event A has occurred, P(A) is the prior probability of event A occurring (i.e., before any evidence is considered), and P(B) is the probability of event B occurring (i.e., the total probability of all possible ways in which event B can occur).

The numerator, P(B|A) * P(A), represents the joint probability of both A and B occurring, while the denominator, P(B), is a normalization factor that ensures that the probabilities add up to 1.

In Bayesian statistics, the **posterior** probability is the probability of a hypothesis or event given the observed data or evidence. It is calculated using Bayes' Theorem, which takes into account both the prior probability of the hypothesis and the likelihood of the data given the hypothesis.

The posterior probability is the updated probability of a hypothesis after taking into account the evidence. It is obtained by multiplying the prior probability by the likelihood function and normalizing the result. The posterior probability can be used to make predictions or decisions based on the observed data.

The **likelihood** function, on the other hand, describes the probability of observing the data given a particular hypothesis or model. It is the key component of Bayes' Theorem, as it allows us to update our beliefs about the hypothesis in light of new evidence.

The likelihood function is often represented by a probability distribution, which assigns a probability to each possible value of the observed data. The shape of the distribution depends on the nature of the data and the underlying hypothesis. For example, if the data are normally distributed, the likelihood function will be a normal distribution centered on the mean of the data.

In Bayesian inference, the likelihood function is used to compute the likelihood ratio, which is the ratio of the likelihoods of two competing hypotheses. The likelihood ratio provides a measure of the strength of the evidence in favor of one hypothesis over the other. The larger the likelihood ratio, the stronger the evidence in favor of the hypothesis.

In PCA, the data is transformed into a new space that is orthogonal to the original space, where the features are ranked by their contribution to the overall variance of the data.

On the other hand, **Word2Vec** is a neural network-based approach for generating vector representations of words in a corpus. It is used to capture the semantic and syntactic meaning of words and their relationships with other words in a corpus. Word2Vec uses a shallow neural network to learn the vector representations of words based on their co-occurrence in a corpus. The resulting vectors are dense and represent the meaning of words in a high-dimensional space. These vector representations can be used for various NLP tasks, such as text classification, information retrieval, and machine translation.

In summary, while both TF-IDF and Word2Vec deal with textual data, they are fundamentally different in their approach and application. TF-IDF is a statistical technique that measures the importance of words in a document or corpus, while Word2Vec is a neural network-based approach for generating vector representations of words in a corpus that capture their semantic and syntactic meaning.

### **8. What is Central Limit Theorem? What general conditions must be satisfied for the central limit theorem to hold?**

The **Central Limit Theorem** (CLT) is a powerful statistical concept that helps us make generalizations about a population based on a sample. The theorem states that the sampling distribution of the mean of any independent, random variables will be approximately normally distributed, with a mean equal to the population mean and a variance equal to the population variance divided by the sample size.

In other words, if we take multiple samples of the same size from a population, the distribution of the means of these samples will approximate a normal distribution, even if the underlying distribution of the population is not normal. This is true for any sample size, provided the sample is sufficiently large (typically, a sample size of at least 30 is considered sufficient).

The CLT is important because it allows us to use the normal distribution to make inferences about the population mean, even if we do not know the population distribution. For example, if we are interested in estimating the mean height of a population, we can take a random sample of people and use the CLT to construct a confidence interval for the population mean.

The CLT has many practical applications in fields such as finance, engineering, and social sciences. It is used to analyze data, make predictions, and estimate parameters in a wide range of contexts. However, it is important to note that the CLT assumes certain conditions, such as independence and identical distribution of the random variables. Violations of these conditions can lead to biased or inaccurate results.

* Sample size: The sample size should be sufficiently large. Typically, a sample size of at least 30 is considered sufficient for the CLT to hold. However, the specific sample size required may depend on the distribution of the underlying population.

### **9. What is N-gram Language Model? What is the difference between Unigram, Bigram and Trigram in NLP? How do N-gram Language Models work? Also mention its applications.**

**N-gram language model:**
An N-gram language model is a statistical model used in natural language processing that predicts the likelihood of a sequence of words based on a probability distribution over the sequence of words. The "N" in N-gram refers to the number of words in a sequence, or "gram." An N-gram model works by estimating the probability of a word given its context, which is typically the N-1 words that precede it. The model estimates the probabilities of each possible N-gram based on a training corpus of text and uses this information to generate likely sentences or to calculate the likelihood of a given sentence.

**Unigram, Bigram, and Trigram:**

Unigram, bigram, and trigram refer to the size of the sequence of words considered by the N-gram model.

* Unigram: A unigram model considers each word in a sentence as a separate entity and does not take into account the context of the other words in the sentence. This means that a unigram model is not able to capture dependencies between words in a sentence, which can limit its effectiveness in certain applications.

* Bigram: A bigram model considers pairs of adjacent words in a sentence. For example, in the sentence "the cat sat on the mat," the bigram model would consider the pairs "the cat," "cat sat," "sat on," "on the," and "the mat."

* Trigram: A trigram model considers sequences of three words at a time. For example, in the sentence "the cat sat on the mat," the trigram model would consider the sequences "the cat sat," "cat sat on," "sat on the," "on the mat."

The choice of N for an N-gram model depends on the particular application and the amount of training data available. A higher value of N can capture more complex dependencies between words but may require more training data to achieve good performance.

**How N-gram Language Models Work:**
N-gram language models work by estimating the probability of a word given its context, which is typically the N-1 words that precede it. The model uses a training corpus of text to estimate the probabilities of each possible N-gram. These probabilities are then used to generate likely sentences or to calculate the likelihood of a given sentence.

The basic formula for an N-gram model is:

P(w_n | w_1, w_2, ..., w_{n-1}) = P(w_1, w_2, ..., w_n) / P(w_1, w_2, ..., w_{n-1})

where P(w_n | w_1, w_2, ..., w_{n-1}) is the probability of the word w_n given the preceding N-1 words, and P(w_1, w_2, ..., w_n) is the probability of the entire N-gram.

N-gram models can be trained using various algorithms, including maximum likelihood estimation and Bayesian methods. In practice, smoothing techniques are often used to handle unseen N-grams and improve the performance of the model.

**Applications of N-gram language models:**

* Speech recognition: N-gram models are commonly used in speech recognition systems to predict the likelihood of a particular sequence of spoken words. For example, if a person says "I want to order a pepperoni pizza," the speech recognition system might use an N-gram model to predict the next word in the sentence, which could be "with" or "without" depending on the context.

* Machine translation: N-gram models are used in machine translation systems to generate likely translations of a given sentence. For example, an N-gram model could be used to predict the next word in a sentence in the target language given the previous words in the source language.

* Spell checking: N-gram models can be used to identify and correct spelling errors by comparing the probability of a given word with its surrounding context. The Gini impurity measures the probability of misclassification of a randomly chosen example from a set. The entropy measure is similar to the Gini impurity measure but considers the information entropy of the set. The misclassification error counts the number of examples that are misclassified in a set. The split quality measure for a given variable is calculated by weighting the split quality of the resulting child nodes by the proportion of examples in each child node. ### **11. What are the different types of gradient descent algorithm? Differentiate  between batch, mini batch and stochastic gradient descent.**

Gradient descent is a widely used optimization algorithm in machine learning that seeks to minimize the cost function of a model. There are several types of gradient descent algorithms, including:

* Batch Gradient Descent: This is the most common type of gradient descent algorithm. In batch gradient descent, the cost function is minimized by computing the gradient of the cost function with respect to the parameters for the entire training set.

* Stochastic Gradient Descent: In this variant of gradient descent, the model parameters are updated after each training example. This approach is useful for large datasets and noisy data.

* Mini-Batch Gradient Descent: Mini-batch gradient descent is a combination of batch and stochastic gradient descent. In this approach, the gradient is computed over a small subset of the training data, known as a mini-batch.

* Conjugate Gradient Descent: This algorithm uses conjugate vectors to efficiently compute the minimum of a quadratic function.

* Momentum-Based Gradient Descent: This algorithm uses a moving average of the gradients to accelerate learning and reduce oscillations.

* Nesterov Accelerated Gradient Descent: This algorithm is a variant of momentum-based gradient descent that uses a modified gradient to compute the update.

* Adagrad: This algorithm adapts the learning rate based on the gradient of the parameters, allowing for larger steps in parameter space for infrequent parameters.

* RMSProp: This algorithm also adapts the learning rate based on the gradient, but with a moving average of the squared gradient to reduce oscillations.

* Adam: This algorithm combines the features of both momentum-based gradient descent and RMSProp to provide faster convergence and better performance.

**Batch Gradient Descent, Mini-Batch Gradient Descent, and Stochastic Gradient Descent** are all optimization algorithms used in machine learning for gradient-based optimization. Here are the main differences between these three algorithms:

1. Batch Gradient Descent: In batch gradient descent, the model parameters are updated based on the average gradient of the entire training dataset. That is, the cost function is computed over the entire dataset, and the gradient of the cost function is calculated with respect to the model parameters. Then, the parameters are updated based on this average gradient. This approach is computationally expensive and can take a long time to converge, but it can lead to a more accurate solution.

2. Mini-Batch Gradient Descent: Mini-batch gradient descent is a variant of batch gradient descent that computes the gradient of the cost function over a small subset of the training data. That is, instead of computing the gradient over the entire dataset, the gradient is computed over a randomly selected subset of the data (usually between 10 and 1,000 examples). This approach is faster than batch gradient descent and can converge more quickly, but it may result in a slightly less accurate solution.

3. Stochastic Gradient Descent: In stochastic gradient descent, the model parameters are updated based on the gradient of the cost function with respect to a single training example. That is, the cost function is computed for each training example in turn, and the model parameters are updated based on the gradient of the cost function for that example. This approach is the fastest but can be the least accurate because the updates can be noisy and may not converge to the global minimum.

In summary, batch gradient descent computes the average gradient over the entire dataset, mini-batch gradient descent computes the gradient over a subset of the data, and stochastic gradient descent computes the gradient over a single training example.

### **12. What is activation function in neural networks and what are its types? What is the difference between Sigmoid, Relu and Softmax activation fuction and which one is better?**

In neural networks, an activation function is a mathematical function applied to the output of each neuron (or node) in a network. Its purpose is to introduce non-linearity into the network, allowing it to learn and model more complex and non-linear relationships between inputs and outputs.

There are several types of activation functions used in neural networks. Here are some of the most common ones:

* Sigmoid function: This function maps any input value to a value between 0 and 1. It is often used in the output layer of a binary classification problem, where the goal is to predict a binary outcome (e.g. yes/no, true/false).

* ReLU function: The Rectified Linear Unit (ReLU) function returns the input value if it is positive, and 0 otherwise. It is widely used in hidden layers of deep neural networks, as it is computationally efficient and has been shown to be effective in many applications.

* Tanh function: The hyperbolic tangent function (tanh) maps any input value to a value between -1 and 1. It is often used in the hidden layers of neural networks, as it is similar to the sigmoid function but with a wider range of output values.

* Softmax function: The softmax function is used in the output layer of a neural network for multi-class classification problems. It maps the outputs of the network to a probability distribution over the possible classes.

There are also other types of activation functions such as the **Leaky ReLU, ELU, and Swish**, among others, which have been proposed to improve the performance of neural networks in certain scenarios.

As discussed earlier The Sigmoid, ReLU, and Softmax activation functions are all commonly used in neural networks, but they have different properties and are suitable for different types of problems.

**The Sigmoid function** is often used in the output layer of a binary classification problem, where the goal is to predict a binary outcome (e.g. yes/no, true/false). It maps any input value to a value between 0 and 1, which can be interpreted as the probability of the input belonging to the positive class. However, the Sigmoid function suffers from the problem of vanishing gradients, where the gradient becomes very small as the input value becomes very large or very small, which can make training the network difficult.

**The ReLU function** is widely used in hidden layers of deep neural networks, as it is computationally efficient and has been shown to be effective in many applications. It returns the input value if it is positive, and 0 otherwise. The ReLU function does not suffer from the vanishing gradients problem, and it has been shown to improve the convergence and training time of deep neural networks.

**The Softmax function** is used in the output layer of a neural network for multi-class classification problems. It maps the outputs of the network to a probability distribution over the possible classes. The Softmax function is similar to the Sigmoid function, but it can handle multiple classes instead of just two. It ensures that the output values add up to 1, which makes it easier to interpret the output as probabilities.

Which activation function is better depends on the problem you are trying to solve and the architecture of your neural network. In general, the ReLU function is a good default choice for hidden layers, while the Sigmoid or Softmax functions are more appropriate for the output layer depending on whether the problem is binary or multi-class classification. However, there are many variations and combinations of activation functions that can be used to achieve better performance, and the choice of activation function should be made based on empirical testing and experimentation.

### **13. What is the difference between Collaborative and Content based Recommender Systems?**

**Collaborative Recommender Systems**:

* Collaborative Filtering (CF) is the most common type of Collaborative Recommender Systems. It involves analyzing user behavior (such as ratings or purchase history) to find similarities and patterns among users. The system can then use this information to recommend items to users with similar behavior.

* CF can be further divided into two types: User-based CF and Item-based CF. User-based CF involves finding similar users to the target user and recommending items that those similar users have liked. Item-based CF, on the other hand, involves finding similar items to the target item and recommending those similar items to the user.

* CF suffers from the cold-start problem, where new users or items have little or no data available for analysis. This makes it difficult to find similar users or items and can result in poor recommendations for new users or items.

**Content-based Recommender Systems**:

* Content-based Recommender Systems analyze the attributes or features of an item (such as the genre, director, actors, etc.) and create a profile for the user based on their past interactions with those attributes.

* However, Content-based Recommender Systems suffer from the problem of overspecialization, where the recommendations become too similar to the user's past interactions and fail to suggest new or diverse items that the user might also like.

* To overcome this limitation, Hybrid Recommender Systems are often used, which combine Collaborative and Content-based approaches to provide more accurate and diverse recommendations.

### **15. What is the difference between covariance and correlation? What are its range and domain for both?**

Covariance and correlation are both measures of the relationship between two variables. However, they differ in the way they are calculated and interpreted.

**Covariance** measures the degree to which two variables vary together. It is a measure of how much one variable changes when the other variable changes. The formula for covariance is:

cov(X,Y) = E[(X - E[X]) * (Y - E[Y])]

where X and Y are the two variables, E[X] and E[Y] are their respective means.

Covariance can take on any value, positive, negative or zero. The sign of the covariance indicates the direction of the relationship between the two variables, positive covariance indicates that the two variables tend to move in the same direction, while negative covariance indicates that they tend to move in opposite directions. The magnitude of the covariance does not have a standard range.

**Correlation**, on the other hand, measures the degree to which two variables are related in a linear way. It is a standardized measure, which means it is independent of the scale of measurement of the variables. The formula for correlation is:

corr(X,Y) = cov(X,Y) / (sd(X) * sd(Y))

where sd(X) and sd(Y) are the standard deviations of X and Y, respectively.

Correlation ranges from -1 to 1, where -1 indicates a perfect negative correlation (i.e., as one variable increases, the other decreases), 0 indicates no correlation, and 1 indicates a perfect positive correlation (i.e., as one variable increases, so does the other). The domain for both covariance and correlation is the set of all pairs of values (x,y) where x and y are the values taken by the two variables being measured.
