### **1.What are hyperparameters? Give some examples. What is hyperparameter tuning?**

In machine learning, hyperparameters are parameters that are not learned from the data, but instead are set by the user prior to training the model. They control the behavior of the model, such as the complexity of the model, the learning rate, and the regularization strength.Moreover, in simple words A hyperparameter is a parameter that is set by the user prior to training a machine learning model, and it controls the behavior of the model during training

Some examples of hyperparameters are:

1. Learning rate: The learning rate controls how quickly the model learns from the data. A high learning rate may cause the model to overshoot the optimal solution, while a low learning rate may cause the model to converge too slowly.

2. Regularization strength: Regularization is used to prevent overfitting of the model to the training data. The regularization strength controls the amount of regularization applied to the model.

3. Number of hidden layers: The number of hidden layers in a neural network is a hyperparameter that determines the complexity of the model. A deeper network with more hidden layers may be able to capture more complex relationships in the data, but it may also be more prone to overfitting.

4. Batch size: The batch size is the number of samples used in each iteration of the training process. A smaller batch size may lead to faster convergence, but it can also make the training process more unstable.

5. Number of epochs: The number of epochs is the number of times the model will cycle through the entire training dataset. Setting the number of epochs too high can lead to overfitting, while setting it too low can lead to underfitting

Hyperparameter tuning, also known as hyperparameter optimization, is the process of selecting the optimal set of hyperparameters for a machine learning model. This involves trying out different combinations of hyperparameters and evaluating their performance on a validation set. The goal is to find the set of hyperparameters that results in the best performance on the validation set. Grid search, random search, and Bayesian optimization are some common techniques used for hyperparameter tuning

### **3. If you move the classification threshold from 0.3 to 0.5 , Will prescision/recall increase or decrease and why?**

Moving the classification threshold from 0.3 to 0.5 will typically result in an increase in precision and a decrease in recall.

Precision is the fraction of true positives among the total number of predicted positives. When the classification threshold is increased from 0.3 to 0.5, the model will be more conservative in predicting positive instances, which means it will be less likely to make false positive predictions. This should increase precision, as the model will be more certain that the predicted positives are true positives.

However, this also means that some true positive instances may be missed, resulting in a decrease in recall. Recall is the fraction of true positives among the total number of actual positives. When the classification threshold is increased, some true positive instances that were previously above the threshold may now be below it, resulting in them being classified as negative.

Overall, the effect on precision and recall will depend on the specific characteristics of the data and the model. In general, increasing the classification threshold will result in a trade-off between precision and recall, with precision increasing and recall decreasing.

### **5. How to find the right threshold for classification problem?**

Finding the right threshold for a classification problem is a crucial step in the modeling process, as it can greatly impact the performance of the model. The threshold determines the probability value above which a predicted class label is considered positive, and below which it is considered negative.

Here are some common methods for finding the right threshold for a classification problem:

1. Default threshold: The default threshold for most classifiers is 0.5. However, this threshold may not be optimal for all problems and data sets.

2. ROC curve: The Receiver Operating Characteristic (ROC) curve is a plot of the True Positive Rate (TPR) against the False Positive Rate (FPR) for different threshold values. The area under the ROC curve (AUC) can be used as a metric to evaluate the performance of different threshold values. Generally, a higher AUC indicates a better performance. The optimal threshold is the one that corresponds to the point on the ROC curve closest to the top-left corner.

3. Precision-Recall curve: The Precision-Recall (PR) curve is another useful tool for evaluating the performance of a binary classifier. The PR curve plots the Precision against the Recall for different threshold values. The optimal threshold is the one that maximizes the F1 score, which is the harmonic mean of Precision and Recall.

4. Cost-sensitive threshold: In some cases, misclassification of one class may be more costly than misclassification of the other class. In such cases, the optimal threshold can be chosen based on the costs associated with misclassification of each class.

5. Domain knowledge: In some cases, domain knowledge may be used to determine an appropriate threshold. For example, if the problem involves identifying fraudulent transactions, a higher threshold may be used to minimize false positives.

Ultimately, the choice of threshold will depend on the specific problem and data set. It may be necessary to experiment with different threshold values and evaluate their impact on performance before selecting the optimal threshold.

Additional to these techniques there is another technique which keeps the escence of both the techniques known as Elastic-Net. A brief description about all the three teqniques:

  * Lasso Regression:

  * Ridge Regression:
Ridge Regression, also known as L2 regularization, adds a penalty term that is proportional to the square of the coefficients of the model.   * Elastic Net:
Elastic Net is a hybrid of Lasso and Ridge regression techniques. It combines the L1 and L2 regularization techniques and adds a penalty term that is a linear combination of the two. 5. Batch normalization: Batch normalization is a technique that normalizes the inputs to each layer in the neural network. This helps prevent overfitting by reducing the internal covariate shift and improving the stability of the model.

These techniques are often used in combination to achieve the best results.

### **8. What is the difference between eucliden distance and Manhattan distance?**

Euclidean distance and Manhattan distance are both measures of distance between two points in a multi-dimensional space. The main difference between them lies in how they calculate this distance.

Euclidean distance, also known as L2 distance, is calculated as the square root of the sum of the squared differences between corresponding coordinates. In other words, it measures the straight-line distance between two points. It is given by the formula:

d(x,y) = sqrt( (x1-y1)^2 + (x2-y2)^2 + ... + (xn-yn)^2 )

where x and y are two n-dimensional vectors.

On the other hand, Manhattan distance, also known as L1 distance or city block distance, is calculated as the sum of the absolute differences between corresponding coordinates. In other words, it measures the distance between two points if one can only travel along the axes, i.e., along the grid-like pattern of a city block. It is given by the formula:

d(x,y) = |x1-y1| + |x2-y2| + ... + |xn-yn|

where x and y are two n-dimensional vectors.

In general, Euclidean distance tends to be more appropriate when the data points have continuous variables and are not sparse, while Manhattan distance is more appropriate when the data points have categorical variables or are sparse. However, the choice between the two ultimately depends on the specific problem and the nature of the data being analyzed.

### **09. Explain the functioning of KNN algorithm. Why is it reffered to as a lazy learner?**

KNN algorithm works by finding the K closest data points (neighbors) to the input data point in the training set, based on some distance metric. The distance metric is usually Euclidean distance, which is calculated by taking the square root of the sum of the squared differences between the corresponding features of the two data points.

Once the K nearest neighbors are identified, the algorithm then classifies the input data point based on the majority class of its nearest neighbors. For example, if the majority of the K neighbors belong to class A, the input data point is classified as belonging to class A as well.

KNN is a simple algorithm that does not require any assumptions about the underlying data distribution. It can be used for both classification and regression tasks, and can handle both numerical and categorical data.

However, KNN is considered a **lazy learner** because it does not learn a model from the training data. Instead, it simply memorizes the training set, and uses this information to make predictions. This makes the algorithm computationally expensive, especially when dealing with large datasets.

Additonally, Here are the steps involved in the KNN algorithm:

1. Load the dataset: First, we need to load the dataset that we want to train the KNN algorithm on.

2. Choose the value of k: Next, we need to choose the value of k, which is the number of nearest neighbors to consider.

3. Calculate distances: Calculate the distance between the new data point and all other data points in the dataset using a distance metric, typically Euclidean distance.

4. Sort distances: Sort the distances in ascending order, so the nearest neighbors are at the top of the list.

5. Select neighbors: Select the "k" nearest neighbors from the sorted distance list.

6. Make a prediction: For classification tasks, take a majority vote from the class labels of the nearest neighbors to make a prediction. For regression tasks, calculate the average of the target values of the nearest neighbors to make a prediction.

7. Evaluate the model: Finally, evaluate the performance of the KNN model using metrics such as accuracy, precision, recall, and F1-score for classification tasks, and mean squared error or R-squared for regression tasks.

### **10. What do you mean by model explainability? How would you introduce explainability to a black box model?**

Model explainability refers to the ability of a machine learning model to provide interpretable explanations of its predictions or decisions. In other words, it's the ability to understand and explain how the model works and why it produced a particular output given a certain input.

Model explainability is crucial in various industries, including finance, healthcare, and law enforcement, where the decisions made by a model can have significant consequences. The lack of explainability in models can lead to mistrust, legal and ethical concerns, and poor adoption rates of in critical applications. Therefore, it is essential to have transparent and explainable models that can provide clear and interpretable insights into their decision-making processes.

A black box model is a model that makes predictions without revealing how it arrived at those predictions, essentially making it difficult to understand its decision-making process. Introducing explainability to a black box model can be done through various methods, including:

1. Feature importance analysis: This involves identifying the most important features that the model used to make its predictions. This can be done through techniques such as permutation importance or SHAP values.

2. Partial dependence plots: This involves plotting the relationship between a specific feature and the model's prediction while holding all other features constant.

3. LIME (Local Interpretable Model-Agnostic Explanations): This involves building a simpler, interpretable model around the black box model for a specific prediction to understand how that prediction was made.

4. Integrated Gradients: This involves computing the contribution of each feature to the final prediction by integrating the gradients of the model output with respect to each feature.

5. Counterfactual explanations: This involves identifying alternative scenarios where the model would have made a different prediction, and explaining why those alternative scenarios didn't happen.

These techniques can help shed light on how a black box model is making its predictions, and make it more interpretable and understandable to stakeholders.

### **11. How do you evaluate a regression problem? What is the difference between R2 and Adjusted R2?**

To evaluate a regression problem, we typically use some performance metrics to assess how well our model fits the data. Some common metrics for regression problems include:

1. Mean Squared Error (MSE): This measures the average squared difference between the predicted and actual values. The lower the MSE, the better the model's performance.

2. Root Mean Squared Error (RMSE): This is the square root of the MSE and is used to measure the average distance between the predicted and actual values. The lower the RMSE, the better the model's performance.

3. R-squared (R2): This measures the proportion of the variance in the dependent variable that is explained by the independent variables. It ranges from 0 to 1, where 1 indicates a perfect fit and 0 indicates no fit.

4. Adjusted R-squared (Adjusted R2): This is a modified version of R2 that takes into account the number of independent variables in the model. It penalizes models with too many variables that may not contribute significantly to the fit.

**The main difference between R2 and Adjusted R2** is that R2 will always increase when more independent variables are added to the model, even if they do not contribute significantly to the fit. Adjusted R2, on the other hand, will only increase if the new variable improves the fit significantly. Therefore, Adjusted R2 is considered a better metric when evaluating models with a large number of independent variables.

In conclusion, while machine learning and deep learning are related fields, they have distinct differences in terms of their algorithms and applications. Machine learning encompasses a broader range of techniques for teaching computers to learn from data, while deep learning specifically involves training artificial neural networks with multiple layers to learn features and representations of data at different levels of abstraction.

### **14. What is vanishing and exploding gradient?**

Vanishing gradient and exploding gradient are both problems that can occur during the training of deep neural networks.

**Vanishing gradient** refers to the situation where the gradient (i.e., the derivative of the loss function with respect to the network's parameters) becomes very small as it is propagated backward through the layers of the network during training. This can happen when the weights in the network are initialized to small values, causing the gradients to also be small, and/or when the network has many layers. As a result, the early layers in the network may not receive meaningful updates during backpropagation and may not learn useful representations.

**Exploding gradient**, on the other hand, occurs when the gradient becomes very large as it is propagated backward through the layers of the network. This can happen when the weights in the network are initialized to large values, causing the gradients to also be large, and/or when the network has many layers. In extreme cases, the gradient can become so large that it causes the weights to update in a way that is too large and destabilizes the training process.

Both of these issues can lead to slow or unstable convergence during training and can make it difficult to train deep neural networks effectively. Various techniques have been developed to address these problems, including weight initialization strategies, regularization techniques, and architectures that can facilitate the flow of gradients through deep networks.

### **15.  What is the difference between ANN and CNN? Why CNN is used for images processing?**

ANN and CNN are both types of artificial neural networks, but they differ in their architecture and the types of problems they are best suited for.

**ANN (Artificial Neural Networks)** are fully connected networks that consist of multiple layers of nodes or neurons. Each neuron receives inputs from all the neurons in the previous layer, and it produces an output that is then passed on to all the neurons in the next layer. ANN can be used for a wide range of applications, including classification, regression, and pattern recognition.

On the other hand, **CNN (Convolutional Neural Networks)** are designed specifically for image processing tasks. The key difference between ANN and CNN is that CNN uses a convolutional layer to scan the image for patterns, rather than fully connected layers. The convolutional layer applies a set of filters to the image, each of which looks for a specific feature, such as edges or corners. The output of the convolutional layer is then passed to a pooling layer, which reduces the size of the feature map and helps to prevent overfitting.
