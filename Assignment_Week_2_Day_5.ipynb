## 1. Give some examples of some random sampling techniques.

Random sampling is a technique used in statistics to select a subset of individuals or items from a larger population in a way that every member of the population has an equal chance of being chosen. Here are some examples of random sampling techniques:

1. Simple random sampling: This involves selecting individuals or items from the population at random using a random number generator or a table of random numbers.

2. Systematic random sampling: This involves selecting every nth individual or item from the population, where n is a predetermined number.

3. Stratified random sampling: This involves dividing the population into subgroups based on a characteristic (e.g., age, gender, income), and then randomly selecting individuals or items from each subgroup in proportion to their size.

4. Cluster random sampling: This involves dividing the population into clusters (e.g., neighborhoods, schools), and then randomly selecting clusters to sample from.

5. Multi-stage random sampling: This involves using a combination of different sampling techniques to select a sample. For example, a researcher might first use cluster sampling to select a group of schools, and then use stratified sampling to select students from each school.

6. Random digit dialing: This is a technique used in telephone surveys where phone numbers are randomly generated and dialed to select participants.

7. Quota sampling: This involves selecting participants based on pre-determined quotas for certain characteristics, such as age, gender, or race.

8. Snowball sampling: This is a technique where participants are recruited through referrals from existing participants.

9. Convenience sampling: This involves selecting participants who are easily accessible or available, such as people passing by on the street or customers at a store.

10. Random sampling with replacement: This technique involves selecting individuals or items from a population randomly and allowing the same individual/item to be selected more than once.

11. Random sampling without replacement: This technique involves selecting individuals or items from a population randomly and not allowing the same individual/item to be selected more than once.

Each of these techniques has its own strengths and weaknesses and may be appropriate for different research questions and study designs

**a)** The Central Limit Theorem is a fundamental result in probability theory that describes the behavior of the sum or average of a large number of independent, identically distributed random variables. It states that under certain conditions, the sum or average of these variables tends to follow a normal distribution, regardless of the distribution of the individual variables.

The conditions required for the Central Limit Theorem to hold are:

1. The random variables must be independent of each other.
2. The random variables must have the same distribution.
3. The random variables must have a finite mean and variance.

The theorem is important because it allows us to make statistical inferences about the mean or sum of a large population based on a small sample, even if we don't know the exact distribution of the population. It is widely used in statistics, econometrics, finance, and other fields where large datasets are analyzed.

More specifically The theorem states that the sum or average of a large number of independent and identically distributed (IID) random variables will tend to be normally distributed, regardless of the underlying distribution of the variables themselves.

To understand the CLT in depth, it's important to first understand what is meant by an IID random variable. An IID random variable is one where each observation is independent of all other observations and is drawn from the same distribution. This means that the probability distribution of each observation is the same, and the values of the observations do not depend on each other.

The CLT says that if we take a large enough sample of IID random variables, the sample mean or sum will tend to be normally distributed with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size. This means that we can use the normal distribution to estimate the probabilities associated with the sample mean or sum, even if we don't know the exact distribution of the population.

The CLT is used in many areas of statistics, including hypothesis testing, confidence interval estimation, and model fitting. It is also used in many practical applications, such as quality control in manufacturing, risk analysis in finance, and election forecasting in politics. In short, the CLT is a powerful and essential tool for statisticians and data scientists alike, and is critical to the understanding of many statistical concepts and techniques.

**b)** The confidence level and significance level are both important concepts in statistical inference, but they are not the same thing.

The significance level (often denoted by the symbol alpha, α) is a measure of how likely it is that the null hypothesis will be rejected when it is actually true. In other words, it is the probability of observing a test statistic as extreme as the one calculated from the sample, assuming that the null hypothesis is true. Typically, a significance level of 0.05 (or 5%) is used as a threshold for rejecting the null hypothesis, meaning that if the p-value of a test is less than 0.05, we reject the null hypothesis.

The confidence level, on the other hand, is a measure of the degree of certainty with which we can say that a particular interval contains the true population parameter. For example, if we construct a 95% confidence interval for the population mean, we can say that we are 95% confident that the true population mean falls within that interval. This means that if we were to repeat our sampling and interval estimation many times, we would expect that 95% of the intervals we construct would contain the true population mean.

The relationship between the confidence level and the significance level is indirect. Typically, a higher confidence level requires a wider interval, which in turn makes it more difficult to reject the null hypothesis. In other words, if we construct a 99% confidence interval, it will be wider than a 95% confidence interval, which means that we will need a stronger signal (a smaller p-value) to reject the null hypothesis at the 99% level than at the 95% level. Therefore, a higher confidence level corresponds to a lower significance level (i.e., a more stringent criterion for rejecting the null hypothesis)

## 1. What is “K” in the KNN Algorithm, how to choose the optimal value of K in the KNN Algorithm, and why is the odd value of “K” preferred over even values in the KNN Algorithm?

**i)** In the KNN (K-Nearest Neighbors) algorithm, "K" represents the number of nearest neighbors that are considered to predict the class of a new data point.

image.png

**ii)** Choosing the optimal value of K is important because it can affect the accuracy of the KNN model. A low value of K will result in a more flexible and complex model that may overfit the data, while a high value of K will result in a more rigid and simple model that may underfit the data. The optimal value of K should strike a balance between these two extremes.

**iii)** To choose the optimal value of K, one can use techniques such as cross-validation and grid search. In cross-validation, the data is divided into k-folds and the model is trained and tested k times. The average accuracy across the k-folds is used as the evaluation metric for different values of K. Grid search involves testing the model with different values of K and selecting the value that yields the highest accuracy.

**iv)** The odd value of K is preferred over even values in the KNN algorithm because it avoids the possibility of a tie when classifying a new data point. When K is even, there is a possibility of an equal number of nearest neighbors belonging to different classes, resulting in a tie. This tie can be broken arbitrarily, leading to inconsistent predictions. On the other hand, when K is odd, there can never be a tie, resulting in a more reliable prediction.

## 2. How can you relate KNN Algorithm to the Bias-Variance tradeoff?

The KNN algorithm can be related to the bias-variance tradeoff in machine learning.

Bias refers to the error introduced by approximating a real-world problem with a simplified model. In the context of KNN, a low value of K results in a more complex model with low bias but high variance. This is because the model becomes more sensitive to the noise and outliers in the training data, which can result in overfitting.

Variance, on the other hand, refers to the error introduced by the sensitivity of the model to fluctuations in the training data. In the context of KNN, a high value of K results in a simpler model with high bias but low variance. This is because the model considers a larger number of nearest neighbors, leading to a smoother decision boundary and reducing the effects of noise and outliers in the training data.

Therefore, **the choice of K in the KNN algorithm involves a tradeoff between bias and variance**. A low value of K reduces bias but increases variance, while a high value of K reduces variance but increases bias. The optimal value of K balances the tradeoff between these two factors, leading to a model with good generalization performance.

## 3. Why Naive Bayes is called Naive and how does the Naive Bayes classifier work?

**Naive Bayes** is called "naive" because it makes a simplifying assumption that the features in the input data are conditionally independent of each other, given the class. This assumption is often violated in real-world scenarios, but the model still performs surprisingly well in practice.

**The Naive Bayes classifier works** by computing the probability of each class label given the input features using Bayes' theorem. Bayes' theorem states that the probability of a hypothesis (in this case, the class label) given the data (the input features) is proportional to the probability of the data given the hypothesis times the prior probability of the hypothesis.

In the case of Naive Bayes, the probability of the data given the hypothesis is computed assuming that each feature is conditionally independent of the others, given the hypothesis. This allows the model to compute the likelihood of the data given each hypothesis as a product of the probabilities of each individual feature, given the hypothesis.

Once the likelihoods are computed, the model applies Bayes' theorem to compute the posterior probability of each hypothesis (class label) given the data. The hypothesis with the highest posterior probability is chosen as the predicted class label.

The model's parameters (i.e., the prior probabilities and the likelihoods) are typically estimated from a training dataset using maximum likelihood estimation or Bayesian inference.

## 4. What is Naive Bayes Classification and Gaussian Naive Bayes?

**Naive Bayes classification** is a probabilistic machine learning algorithm used for classification tasks. It is based on Bayes' theorem and the assumption of conditional independence between features, given the class label. The algorithm calculates the probability of each class given a set of input features, and then selects the class with the highest probability as the predicted class for the input.

**Gaussian Naive Bayes is a variant of Naive Bayes classification** where the input features are assumed to follow a Gaussian (normal) distribution. This assumption is useful when the input features are continuous variables, and it allows the algorithm to model the distribution of each feature using the mean and variance of the feature values for each class label.

In **Gaussian Naive Bayes**, the algorithm calculates the likelihood of the input features given each class using the probability density function of the normal distribution. The mean and variance of the distribution are estimated from the training data for each class. The algorithm then applies Bayes' theorem to calculate the posterior probability of each class given the input features, and selects the class with the highest posterior probability as the predicted class.

Gaussian Naive Bayes is widely used in various applications, including text classification, image classification, and medical diagnosis, among others. However, it may not work well in cases where the input features do not follow a Gaussian distribution or when there are dependencies between the features. In such cases, other variants of Naive Bayes or other machine learning algorithms may be more suitable.
